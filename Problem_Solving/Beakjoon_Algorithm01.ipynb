# 10828번: 스택
정수를 저장하는 스택을 구현한 다음, 입력으로 주어지는 명령을 처리하는 프로그램을 작성하시오.  

명령은 총 다섯 가지이다.  

push X: 정수 X를 스택에 넣는 연산이다.  
pop: 스택에서 가장 위에 있는 정수를 빼고, 그 수를 출력한다. 만약 스택에 들어있는 정수가 없는 경우에는 -1을 출력한다.  
size: 스택에 들어있는 정수의 개수를 출력한다.  
empty: 스택이 비어있으면 1, 아니면 0을 출력한다.  
top: 스택의 가장 위에 있는 정수를 출력한다. 만약 스택에 들어있는 정수가 없는 경우에는 -1을 출력한다.  

```python
#프린트 버전
N = int(input())
stack = []
prt= []

for i in range(N):
    inputs = input().split()
    
    if len(inputs) == 1:
        s = inputs[0]
        if s == "pop":
            if stack:
                prt.append(len(stack))
            else:
                prt.append(-1)
                #print(-1)
        elif s == "top":
            if stack:
                prt.append(stack[-1])
                #print(stack[-1])
            else:
                prt.append(-1)
                #print(-1)
        elif s == "empty":
            if stack:
                prt.append(0)
                #print(0)
            else:
                prt.append(1)
                #print(1)
        elif s == "size":
            prt.append(len(stack))
            #print(len(stack))
        else:
            print("error")
    elif len(inputs) == 2:
        s, n = inputs
        n = int(n)
        if s == "push":
            stack.append(n)
            #prt.append(stack[-1])
            #print(stack[-1])
        else:
            print("error")
    else:
        print("입력값 갯수를 1개 또는 2개로 하시오")

for i in prt:
    print(i)
```

```python
#최종본
N = int(input())
stack = []
prt= []

for i in range(N):
    inputs = input().split()
    
    if len(inputs) == 1:
        s = inputs[0]
        if s == "pop":
            if stack:
                prt.append(stack[-1])
                stack.pop()
            else:
                prt.append(-1)
        elif s == "top":
            if stack:
                prt.append(stack[-1])
            else:
                prt.append(-1)
        elif s == "empty":
            if stack:
                prt.append(0)
            else:
                prt.append(1)
        elif s == "size":
            prt.append(len(stack))
        else:
            print("error")
    elif len(inputs) == 2:
        s, n = inputs
        n = int(n)
        if s == "push":
            stack.append(n)
        else:
            print("error")
    else:
        print("error")

for i in prt:
    print(i)
```
